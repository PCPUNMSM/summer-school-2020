```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def runge_kutta4(a, b, h, f, y0):
    #h = (b - a)/N
    N = int((b - a)/h)
    
    x = np.linspace(a, b, N + 1)
    
    y = np.zeros(N + 1)
    
    y[0] = y0
    
    #print(x[0], y[0])
    
    for i in range(N):
        k1 = h*f(y[i], x[i])
        k2 = h*f(y[i] + k1/2, x[i] + h/2)
        k3 = h*f(y[i] + k2/2, x[i] + h/2)
        k4 = h*f(y[i] + k3, x[i + 1])
        
        y[i + 1] = y[i] + 1/6*(k1 + 2*k2 + 2*k3 + k4)
        
        #print(x[i + 1], y[i + 1])
        
    return y
```

```python
def f(y, x):
    f0 = 1e-9
    c = 1e-18
    H = 1e-2
    
    f = 2*f0*np.exp(-c*(1 + H**2))*(np.sinh(2*c*H) - y*np.cosh(2*c*H))
    
    return f

a = 0.0
b = 10.0
h = 0.1
y0 = 1.0

N = int((b - a)/h)

x = np.linspace(a, b, N + 1)
y = runge_kutta4(a, b, h, f, y0)
```

```python
plt.plot(x, y)
#plt.plot(xsol, ysol)
#plt.legend(['euler', 'exacto'])
plt.show()
```
