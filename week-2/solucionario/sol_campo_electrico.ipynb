# Campo eléctrico - Solución

Importando los paquetes necesarios.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
from scipy.constants import epsilon_0 as e0

%matplotlib inline
```

Definimos una función para calcular el campo eléctrico y el módulo de un vector.

```python
def E(x, y, q, a, b):
    k = 1/(4*np.pi*e0)
    
    Ex = k*q*(x - a)/((x - a)**2 + (y - b)**2)
    Ey = k*q*(y - b)/((x - a)**2 + (y - b)**2)
    
    return Ex, Ey

def modulo(Ax, Ay):
    A = (Ax**2 + Ay**2)**(1/2)
    return A
```

# Campo eléctrico

```python
l = 4e-2

x = np.linspace(-l - 1e-2, l + 1e-2, 10)
y = np.linspace(-l - 1e-2, l + 1e-2, 10)

X, Y = np.meshgrid(x, y)

q1 = 3e-6
q2 = 4e-6
q3 = -7e-6

Ex1, Ey1 = E(X, Y, q1, -l, -l)
Ex2, Ey2 = E(X, Y, q2, -l, l)
Ex3, Ey3 = E(X, Y, q3, l, -l)

Etx = Ex1 + Ex2 + Ex3
Ety = Ey1 + Ey2 + Ey3

Emod = modulo(Etx, Ety)

plt.quiver(X, Y, Etx/Emod, Ety/Emod, Emod, pivot = 'middle', cmap='cool', 
           norm = colors.LogNorm(vmin = Emod.min(),vmax = Emod.max()))

circle1 = plt.Circle((-l, -l), 0.2e-2, color = 'r')
circle2 = plt.Circle((-l, l), 0.2e-2, color = 'r')
circle3 = plt.Circle((l, -l), 0.2e-2, color = 'b')

plt.gca().add_patch(circle1)
plt.gca().add_patch(circle2)
plt.gca().add_patch(circle3)
plt.axis('scaled')

plt.xticks([-4e-2, -2e-2, 0, 2e-2, 4e-2], ['-4', '-2', '0', '2', '4'])
plt.yticks([-4e-2, -2e-2, 0, 2e-2, 4e-2], ['-4', '-2', '0', '2', '4'])
plt.xlabel('x (cm)')
plt.ylabel('y (cm)')

plt.show()
```

## Líneas de campo

```python
plt.streamplot(X, Y, Etx, Ety, color = Emod, cmap='cool', norm = colors.LogNorm(vmin = Emod.min(),vmax = Emod.max()), arrowstyle='->')
plt.axis('scaled')
plt.xlim([-l - 1e-2, l + 1e-2])
plt.ylim([-l - 1e-2, l + 1e-2])
plt.show()
```

## Potencial eléctrico

```python
def V(x, y, q, a, b):
    k = 1/(4*np.pi*e0)
    V = k*q/((x - a)**2 + (y - b)**2)**(1/2)
    return V
```

```python
V1 = V(X, Y, q1, -l, -l)
V2 = V(X, Y, -q2, -l, l)
V3 = V(X, Y, -q3, l, -l)

Vt = V1 + V2 + V3

plt.contourf(X, Y, Vt, levels = 5, cmap = 'summer')
plt.contour(X, Y, Vt, levels = 5, colors = 'k')
plt.axis('scaled')

circle1 = plt.Circle((-l, -l), 0.2e-2, color = 'r')
circle2 = plt.Circle((-l, l), 0.2e-2, color = 'r')
circle3 = plt.Circle((l, -l), 0.2e-2, color = 'b')

plt.gca().add_patch(circle1)
plt.gca().add_patch(circle2)
plt.gca().add_patch(circle3)
plt.axis('scaled')

plt.xticks([-4e-2, -2e-2, 0, 2e-2, 4e-2], ['-4', '-2', '0', '2', '4'])
plt.yticks([-4e-2, -2e-2, 0, 2e-2, 4e-2], ['-4', '-2', '0', '2', '4'])
plt.xlabel('x (cm)')
plt.ylabel('y (cm)')

plt.show()
```
